```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/workspaces/Nuevo-repo/data/raw/spotify.csv')
data
```

```python
data.shape
```

```python
numero_de_faltantes = data.isna().sum()

celdas = np.product(data.shape)
faltantes_totales = numero_de_faltantes.sum()
porcentaje_faltantes = (faltantes_totales / celdas) * 100
porcentaje_faltantes = round (porcentaje_faltantes,4)
```

```python
print("El porcentaje de datos faltantes en el conjunto de datos es: {}%".format(porcentaje_faltantes))
```

Se puede observar que es un n√∫mero muy bajo el de los datos faltantes.

Codificaci√≥n de etiquetas de variables categ√≥ricas

```python
def labelencoder(df):
    for c in df.columns:
        if df[c].dtype=='object': 
            df[c] = df[c].fillna('N')
            lbl = LabelEncoder()
            lbl.fit(list(df[c].values))
            df[c] = lbl.transform(df[c].values)
    return df
```
