Eliminación de columna innecesaria "Unnamed 0"

```python
df = data.drop("Unnamed: 0", axis=1)
df
```

VALORES NULOS

```python
df.isna().sum()
```

Se muestra que la mayoría de las columnas no tienen valores faltantes, excepto las columnas: "artists", "album_name" y "track_name", que tienen un valor faltante cada una.

```python
df = df.dropna()
```

```python
df.isna().sum()
```

Se eliminaron los pocos valores faltantes con los que se contaba en el dataset.

VALORES DUPLICADOS

```python
df = df.drop_duplicates()
df.shape
```

De esta manera podemos concluir que se han eliminado 451 filas debido a valores duplicados.
