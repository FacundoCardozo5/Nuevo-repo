```python
pd.set_option('display.max_rows', None)
serie=(df.isnull().sum()/df.shape[0])*100
serie=serie.sort_values(ascending=False)
serie= serie[serie>0]
serie
```

Se concluye que el porcentaje de valores nulos de las tres columnas: 'artists', 'album_name' y 'track_name' es muy bajo, alrededor del 0.0877%, lo que sugiere que los datos en estas columnas est√°n completos en su gran mayor√≠a.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
serie.plot(kind='bar')
```
